```
import numpy as np 
import pandas as pd 
import seaborn as sns
sns.set(style="whitegrid", palette="muted")
import matplotlib.pyplot as plt
```

```
data_df=pd.read_csv("Iris.csv")
```

```
data_df.head()
```

```
b=sns.jointplot(x='SepalWidthCm', y='SepalLengthCm', data=data_df, height=8,alpha=.6,color='r', marker='x')
#b=sns.jointplot(x='SepalWidthCm', y='SepalLengthCm', data=data_df)
```

```
c=sns.FacetGrid(data_df,col='Species')
c.map(plt.scatter,'SepalLengthCm','SepalWidthCm')
```

```
e=sns.catplot(x="SepalLengthCm", y="SepalWidthCm",col="Species", data=data_df, kind="swarm");
#e=sns.catplot(x="SepalLengthCm", y="SepalWidthCm", data=data_df, kind="swarm");
e.set(xticklabels=[])
```

```
e=sns.violinplot(x='Species',y='SepalLengthCm',data=data_df)
```

```
iris = sns.load_dataset("iris")
iris.head()
sns.boxplot(x="species", y="petal_length", data=iris, palette="magma")
```

```
sns.swarmplot(x="species", y="petal_length", data=iris)
```

```
iris = sns.load_dataset('iris')

sns.pairplot(iris, hue='species', palette='gist_heat', dropna=True)
#sns.pairplot(iris,hue="species")
```

```
tips = sns.load_dataset("tips")

tips.head(10)
```

```
sns.barplot(x="day", y="tip", data=tips)
```

```
sns.set_style("ticks")
sns.boxplot(x="day", y="total_bill", data=tips)
#sns.despine(left=True)
```

```
sns.lmplot(x="total_bill", y="tip", data=tips)
```

```
sns.lmplot(x="total_bill", y="tip", hue="sex",data=tips)
```

```
sns.lmplot(x="total_bill", y="tip", col="smoker",data=tips)
```

```
sns.lmplot(x="total_bill", y="tip", hue="smoker",col="sex",data=tips, markers=['o','x'], palette="Set1")
```

```
sns.lmplot(x="total_bill", y="tip", row="sex",data=tips, col="time",height=3 , palette="magma")
sns.set_style="whitegrid"
```

```
sns.pairplot(data=tips, hue='sex', palette='Set1', x_vars=['total_bill','size'], y_vars=['tip'], height=6, aspect=.85, kind='reg')
```

```
sns.set(style='darkgrid')
```

```
sns.relplot(x='total_bill', y='tip', data=tips, size='size', sizes=(1,5), style='sex' , kind='line', sort=True)
```
