```python
plt.subplot(2,1,1)  # first plot
plt.plot([x for x in range(25)], [y**2 for y in range(25)])
plt.plot([x for x in range(25)], [y for y in range(25)])
plt.title('Square Numbers')

plt.subplot(2,1,2) # second plot
plt.plot([x for x in range(25)], [y**3 for y in range(25)])
plt.plot([x for x in range(25)], [y for y in range(25)])
plt.title('bhanu of Numbers')

```

```python
# multiple plots

plt.subplot(2,1,1)  # first plot
plt.plot([x for x in range(25)], [y**2 for y in range(25)])
plt.plot([x for x in range(25)], [y for y in range(25)])
plt.title('Square Numbers')

plt.subplot(2,1,2) # second plot
plt.plot([x for x in range(25)], [y**3 for y in range(25)])
plt.plot([x for x in range(25)], [y for y in range(25)])
plt.title('Cube of Numbers')

plt.subplots_adjust(top=1, bottom=0.1, left=0.10, right=0.95, hspace=1,
                    wspace=1)
```
